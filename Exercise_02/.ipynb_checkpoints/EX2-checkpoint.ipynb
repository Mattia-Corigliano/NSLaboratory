### 2.1.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("output.txt").T
data2 = np.loadtxt("output2.txt").T
plt.figure(figsize=(10, 3), dpi=100)
plt.subplot(1, 2, 1); plt.errorbar(data[0], data[1]-1, yerr=data[2], color='b', ecolor="gray", capsize=2)
plt.title("Monte Carlo Method", size = 20)
plt.xlabel("#points", size=16)
plt.ylabel("I-1", size=16)
plt.grid(True)
plt.subplot(1, 2, 2); plt.errorbar(data2[0], data2[1]-1, yerr=data2[2], color='b', ecolor="gray", capsize=2)
plt.title("Importance sampling", size=20)
plt.xlabel("#points", size=16)
plt.ylabel("I-1", size=16)
plt.grid(True)
plt.subplots_adjust(wspace=0.4)
plt.show()
```

### 2.2.

```python
from mpl_toolkits.mplot3d import Axes3D

data3 = np.loadtxt("output3.txt").T
data4 = np.loadtxt("output4.txt").T
data5 = np.loadtxt("output5.txt").T

fig = plt.figure(figsize=(10, 3), dpi=100)
ax = fig.add_subplot(121, projection='3d')
ax.set_title("cubic lattice")
for j in range(1000):
    x = np.zeros(303)
    y = np.zeros(303)
    z = np.zeros(303)

    for i in range(101):
        x[i] = data3[3*i+303*j]
        y[i] = data3[3*i+1+303*j]
        z[i] = data3[3*i+2+303*j]
    ax.plot(x, y, z, "-", lw=0.2)

ax = fig.add_subplot(122, projection='3d')
ax.set_title("Ideal chain 100 monomers")
for j in range(1000):
    x = np.zeros(303)
    y = np.zeros(303)
    z = np.zeros(303)

    for i in range(101):
        x[i] = data4[3*i+303*j]
        y[i] = data4[3*i+1+303*j]
        z[i] = data4[3*i+2+303*j]
    ax.plot(x, y, z, "-", lw=0.25)
plt.savefig("fig2.pdf")
plt.show()

plt.figure(figsize=(10, 3), dpi=100)
plt.subplot(1, 2, 1); plt.errorbar(data5[0], data5[1], yerr=data5[3], fmt = "-", 
                                   color = 'b', ecolor = 'grey', capsize = 2, label="rad<r_N^2>")
plt.plot(data5[0], np.sqrt(data5[0]), 'r-', lw=3, label="fit")
plt.grid(True)
plt.legend()
plt.subplot(1, 2, 2); plt.errorbar(data5[0], data5[2], yerr=data5[4], fmt = "-", 
                                   color = 'b', ecolor = 'grey', capsize = 2, label="rad<r_N^2>")
plt.plot(data5[0], np.sqrt(data5[0]), 'r-', lw=3, label="rad(N)")
plt.grid(True)
plt.legend()

plt.show()
```
