# Numerical Simulation Laboratory

## Notebook 04

### Exercise 04.1

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("plot1.txt").T
data2 = np.loadtxt("plot2.txt").T
data3 = np.loadtxt("plot3.txt").T
data4 = np.loadtxt("plot4.txt").T
data5 = np.loadtxt("plot5.txt").T
nsteps = np.linspace(1, 1000, 1000)

plt.figure(figsize=(5, 4), dpi=100)
plt.plot(nsteps, data, 'b-', lw=2, label="Starting run")
plt.plot(nsteps, data2, 'r-', lw=1, label="first restart")
plt.plot(nsteps, data3, 'y-', lw=1, label="second restart")
plt.plot(nsteps, data4, 'g-', lw=1, label="third restart")
plt.plot(nsteps, data5, 'k-', lw=2, label="fourth restart")
plt.grid(True)
plt.xlabel("number of steps", size=16)
plt.ylabel("temperature [kbT/e]", size=16)
plt.title("solid phase", size=20)
plt.legend()
plt.show()
```

### Exercise 04.2

```python
data6 = np.loadtxt("ave_temp.out").T
data7 = np.loadtxt("ave_ekin.out").T
data8 = np.loadtxt("ave_epot.out").T
data9 = np.loadtxt("ave_etot.out").T

plt.figure(figsize=(10, 8), dpi=100)
plt.subplot(2, 2, 1); plt.errorbar(data6[0], data6[1], yerr=data6[2], color='b', 
                                   lw=2, ecolor="gray", capsize=2, label="temperature")
plt.grid(True)
plt.legend()
plt.subplot(2, 2, 2); plt.errorbar(data7[0], data7[1], yerr=data7[2], color='b', 
                                   lw=2, ecolor="gray", capsize=2, label="kinetic energy per particle")
plt.grid(True)
plt.legend()
plt.subplot(2, 2, 3); plt.errorbar(data8[0], data8[1], yerr=data8[2], color='b', 
                                   lw=2, ecolor="gray", capsize=2, label="potential energy per particle")
plt.grid(True)
plt.legend()
plt.subplot(2, 2, 4); plt.errorbar(data9[0], data9[1], yerr=data9[2], color='b', 
                                   lw=2, ecolor="gray", capsize=2, label="total energy per particle")
plt.grid(True)
plt.legend()
plt.subplots_adjust(wspace=0.3)
plt.subplots_adjust(hspace=0.2)
plt.show()
```

### Exercise 04.3

# Simulating Argon

## Solid Phase

### Equilibrating the system

```python
Argondata = np.loadtxt("Argonplot1.txt").T
Argondata2 = np.loadtxt("Argonplot2.txt").T
Argondata3 = np.loadtxt("Argonplot3.txt").T
Argondata4 = np.loadtxt("Argonplot4.txt").T
Argondata5 = np.loadtxt("Argonplot5.txt").T
nsteps = np.linspace(1, 1000, 1000)

plt.figure(figsize=(5, 4), dpi=100)
plt.plot(nsteps, Argondata, 'b-', lw=2, label="Starting run")
plt.plot(nsteps, Argondata2, 'r-', lw=1, label="first restart")
plt.plot(nsteps, Argondata3, 'y-', lw=1, label="second restart")
plt.plot(nsteps, Argondata4, 'g-', lw=1, label="third restart")
plt.plot(nsteps, Argondata5, 'k-', lw=2, label="fourth restart")
plt.grid(True)
plt.xlabel("number of steps", size=16)
plt.ylabel("temperature [kbT/e]", size=16)
plt.title("solid phase", size=20)
plt.legend()
plt.show()
```

```python
Argondata6 = np.loadtxt("ARGONave_temp.out").T
Argondata7 = np.loadtxt("ARGONave_ekin.out").T
Argondata8 = np.loadtxt("ARGONave_epot.out").T
Argondata9 = np.loadtxt("ARGONave_etot.out").T

plt.figure(figsize=(10, 8), dpi=100)
plt.subplot(2, 2, 1); plt.errorbar(Argondata6[0], 120*Argondata6[1], yerr=120*Argondata6[2], color='b', 
                                   lw=2, ecolor="gray", capsize=2, label="temperature")
plt.grid(True)
plt.legend()
plt.ylabel("Temperature [K]", size=14)
plt.subplot(2, 2, 2); plt.errorbar(Argondata7[0], 1.657*10**(-21)*Argondata7[1], yerr=1.657*10**(-21)*Argondata7[2], color='b', 
                                   lw=2, ecolor="gray", capsize=2, label="kinetic energy per particle")
plt.grid(True)
plt.legend()
plt.ylabel("av. kinetic energy per particle [J]")
plt.subplot(2, 2, 3); plt.errorbar(Argondata8[0], 1.657*10**(-21)*Argondata8[1], yerr=1.657*10**(-21)*Argondata8[2], color='b', 
                                   lw=2, ecolor="gray", capsize=2, label="potential energy per particle")
plt.grid(True)
plt.legend()
plt.ylabel("av. potential energy per particle [J]")
plt.subplot(2, 2, 4); plt.errorbar(Argondata9[0], 1.657*10**(-21)*Argondata9[1], yerr=1.657*10**(-21)*Argondata9[2], color='b', 
                                   lw=2, ecolor="gray", capsize=2, label="total energy per particle")
plt.grid(True)
plt.legend()
plt.ylabel("av. total energy per particle [J]")
plt.subplots_adjust(wspace=0.3)
plt.subplots_adjust(hspace=0.2)
plt.show()
```
