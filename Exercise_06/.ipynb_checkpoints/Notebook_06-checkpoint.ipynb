## Simulation of the 1D Ising Model

*In this computational exercise we simulated the 1D Ising model with the **Metropolis sampling** and the **Boltzmann sampling** algorithm.*

*We also calculated some of its thermodynamics properties using the data-blocking technique.*

*Since the 1D Ising Model is exactly solvable, we compared the results found with our simulations with the analytical curves. We found that both sampling algorithms do a great job in evaluating the thermodynamics properties of the system.*

### Equilibration of the system

As for the Molecular dynamics code, the system has to be equilibrated (with both the sampling algorithms) at the desired temperature before performing measurements. That is, the simulation starts from a defined initial configuration ($T= \infty$ here) and after some Metropolis (or Gibbs) steps it reaches equilibrated configurations with the desired temperature. Once the system is equilibrated we can consider successive configurations as legit sample configurations over which performing our measurements.

In the panel below, the behaviour of the energy per spin as a function of the number of step is displayed.

```python
#T=0.5
Nplots = 3
colors = ['blue', 'red', 'black']
labels = ['starting run', 'first restart', 'second restart']
fig = plt.figure(figsize=(12, 4), dpi=120)
plt.suptitle('Equilibration of the system', size=14)
mpl.rcParams['figure.subplot.top'] = 0.9

for i in range(Nplots):
    data = np.loadtxt("/home/mattia/Scrivania/NSLaboratory/Exercises/Exercise_06/data/T=0.5/eq{}.txt".format(i+1)).T
    plt.plot(data[0]+50000*i, data[1], color = colors[i], label=labels[i], lw = 0.5)
plt.ylabel("Energy per spin")
plt.xlabel("Step")

#T=2
for i in range(Nplots):
    data = np.loadtxt("/home/mattia/Scrivania/NSLaboratory/Exercises/Exercise_06/data/T=2/eq{}.txt".format(i+1)).T
    plt.plot(data[0]+50000*i, data[1], color = colors[i], lw = 0.5)

plt.legend()
plt.show()
```

### Thermodynamics Measures

```python
measures = ['U_T', 'C_T', 'Chi_T', 'Mag_T', 
            'GIBBS_U_T', 'GIBBS_C_T', 'GIBBS_Chi_T', 'GIBBS_Mag_T']
titles = ['internal energy (h=0)', 'Heat capacity (h=0)',
          'Susceptibility (h=0)','magnetization (h = 0.02)']
Nplots = 4

fig = plt.figure(figsize=(12, 9), dpi=120)
plt.suptitle('1D Ising Model', size=14)
mpl.rcParams['figure.subplot.top'] = 0.9

#analytical curves
## energy
points=100
T = np.linspace(0.5,2.0,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )

## Heat capacity
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)

## Susceptibility
X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)

## Magnetization
h=0.02 #external field
b = 1/T
l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)

for k in range(Nplots):
    ## loading data
    data = np.loadtxt("/home/mattia/Scrivania/NSLaboratory/Exercises/Exercise_06/data/measures/{}.txt".format(measures[k])).T
    data2 = np.loadtxt("/home/mattia/Scrivania/NSLaboratory/Exercises/Exercise_06/data/measures/{}.txt".format(measures[k+4])).T  
    plt.subplot(2, 2, k+1);
    if(k==0):
        plt.plot(T, e, 'y-', label='theoretical prediction')
    if(k==1):
        plt.plot(T, heat, 'y-', label='theoretical prediction')
    if(k==2):
        plt.plot(T, X, 'y-', label='theoretical prediction')
    if(k==3):
        plt.plot(T, M, 'y-', label='theoretical prediction')
    plt.errorbar(data[0], data[1],yerr=data[2],color='b', ecolor = 'gray', 
             linestyle = 'none', marker ='D', ms=2, capsize=2, label='metropolis')
    plt.errorbar(data2[0], data2[1],yerr=data2[2],color='r', ecolor = 'gray', 
             linestyle = 'none', marker ='^', ms=2, capsize=2, label='gibbs sampling')
    
    plt.title(titles[k])
    plt.xlabel('T')
    plt.ylabel(measures[k])
    plt.legend()
    

plt.show()
```
