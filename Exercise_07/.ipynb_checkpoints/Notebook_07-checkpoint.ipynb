## Monte Carlo simulation of a Lennard-Jones fluid

*In this computational exercise we tackled the same problem of the 4th computational exercise - i.e. the simulation of a Lennard-Jones fluid - under the different approach of Monte Carlo methods.*

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize

##  setting the Layout
import matplotlib as mpl
## LINES
mpl.rcParams['lines.linewidth'] = 2
mpl.rcParams['lines.linestyle'] = '-'
## AXES
mpl.rcParams['axes.linewidth'] = 1 ## edge linewidth
mpl.rcParams['axes.grid'] = True ## display grid or not
mpl.rcParams['axes.titlesize'] = 16 ## fontsize of the axes title
#mpl.rcParams['axes.titleweight'] = 'bold'
mpl.rcParams['axes.labelsize'] = 14  ## fontsize of the x any y labels
mpl.rcParams['axes.formatter.limits'] = [-2,6] # use scientific notation if log10
mpl.rcParams['axes.formatter.use_mathtext'] = True # When True, use mathtext for scientific
mpl.rcParams['legend.fontsize'] = 12
## TICKS
mpl.rcParams['xtick.labelsize'] = 12 ## fontsize of the tick labels
mpl.rcParams['ytick.labelsize'] = 12 
mpl.rcParams['xtick.minor.visible'] = True ## visibility of minor ticks on x-axis
mpl.rcParams['ytick.minor.visible'] = True
## FIGURE
mpl.rcParams['figure.titleweight'] = 'bold' ## weight of the figure title
mpl.rcParams['figure.subplot.wspace'] = 0.3
mpl.rcParams['figure.subplot.hspace'] = 0.4
## ERRORBARS
mpl.rcParams['errorbar.capsize'] = 2  ## length of end cap on error bars in pixels

# autocorrelation function
def autocorrelation(v,t):
    #denominator
    ave=0
    av2=0
    for i in range(len(v)):
        ave=ave + v[i]
        av2=av2 + v[i]*v[i]
    ave=ave/(len(v))
    av2=av2/(len(v))
    #numerator
    delta_t =len(v) -t # t_max - t
    corr1=0
    corr2=0
    corr3=0
    for i in range(delta_t):
        corr1=corr1+v[i]*v[i+t]
        corr2=corr2+v[i]
        corr3=corr3+v[i+t]
    ac = corr1/delta_t-(corr2/delta_t)*(corr3/delta_t)
    ac = ac/(av2-ave**2)

    return ac

#exponential fit
def exp_fit(x,a):
    return np.exp(-a*x)

#datablocking error
def datablocking_error(v, N):
    M = len(v)
    L = int(M/N)
    measure=np.zeros(int(N))
    measure2=np.zeros(int(N))
    vblock= np.zeros(L)
    ave = 0
    ave2=0
    err=0
    for i in range(N):
        if(i==N-1):
            vblock=v[L*i::1]
        else:
            vblock = v[L*i:L*(i+1):1]
        
        measure[i] = np.mean(vblock)
        measure2[i] = measure[i]**2
        ave = ave + measure[i]/N
        ave2 = ave2 + measure2[i]/N
    
    err=np.sqrt((ave2-ave**2)/(N-1))
    return err
```

### A. Equilibration and autocorrelation function computation

In this part of the computational exercise we equilibrated the system and studied in more detail, with the aid of the autocorrelation function, how large should the block size be in the data-blocking method.

### Solid Phase

#### Equilibration


```python
solid_noneq_instant_values = np.loadtxt("data/equilibrations/solid_noneq_instant_values.txt").T
solid_eq_instant_values = np.loadtxt("data/equilibrations/solid_eq_instant_values.txt").T

fig = plt.figure(figsize=(12, 4), dpi=120)
plt.suptitle("Equilibration of the solid phase", size=14)
plt.subplot(1, 2, 1); 
plt.plot(solid_noneq_instant_values[0][::10], solid_noneq_instant_values[1][::10], 'b-', lw=0.5)
plt.plot(solid_eq_instant_values[0][::10]+500001, solid_eq_instant_values[1][::10], 'r-', lw=0.5, label='equilibrated configurations')
plt.semilogx()
plt.xlabel('step')
plt.ylabel('potential energy per particle')
plt.legend()
plt.ylim(-8., -7.)
plt.subplot(1, 2, 2); 
plt.plot(solid_noneq_instant_values[0][::10], solid_noneq_instant_values[2][::10], 'b-', lw=0.5)
plt.plot(solid_eq_instant_values[0][::10]+500001, solid_eq_instant_values[2][::10], 'r-', lw=0.5, label='equilibrated configurations')
plt.semilogx()
plt.ylim(5., 10.)
plt.xlabel('step')
plt.ylabel('Pressure')
plt.legend()
plt.show()
```

#### Autocorrelation function computation

```python
time = np.arange(101)
ac_pot = np.zeros(101)
ac_pres= np.zeros(101)

for t in time :
    ac_pot[t] = autocorrelation(solid_eq_instant_values[1], t)
    ac_pres[t] = autocorrelation(solid_eq_instant_values[2], t)
```

```python
params_pot, cov_pot= optimize.curve_fit(exp_fit, time, ac_pot)  
params_pres, cov_pres= optimize.curve_fit(exp_fit, time, ac_pres)

fig = plt.figure(figsize=(12, 4), dpi=120)
plt.suptitle('Autocorrelation function computation (solid)', size=14)
mpl.rcParams['figure.subplot.top'] = 0.8

plt.subplot(1, 2, 1);
plt.plot(time, ac_pot, color='b', marker='.', markersize=4, linestyle='none')
plt.plot(time, exp_fit(time, params_pot[0]) , 'r-', lw=2, label='exponential fit')
plt.title('potential energy per particle')
plt.xlabel('time')
plt.ylabel('autocorrelation U/N')
plt.legend()
plt.subplot(1, 2, 2);
plt.title('pressure')
plt.plot(time, ac_pres, color='b', marker='.', markersize=4, linestyle='none')
plt.plot(time, exp_fit(time, params_pres[0]) , 'r-', lw=2, label='exponential fit')
plt.xlabel('time')
plt.ylabel('autocorrelation P')
plt.legend()
plt.show()

print("Estimated correlation time for potential energy = "+str(1/params_pot[0]))
print("Estimated correlation time for pressure = "+str(1/params_pres[0]))
```

**NB:**

**How large a block should be?**

With the analysis above we addressed again (see also Ex 1 and Ex 5) the problem of the optimal block size to be used for data blocking.

As shown in the panels "data-blocking error estimation", a general trend of the error as a function of the number of blocks can be inferred:

- there is a kind of plateau region, of about two logs (a little bit larger for gas), consisting of values of the number of blocks at which the error equlibrates around a specific value. As we'll see, we should consider this constant value as the most reliable estimate of the error in our measurement. As a consequence, we should try to setup our blocking method with a number of blocks in this optimal region.

Indeed, 
- before this region, i.e. for less number of blocks (or blocks of bigger size), we might end up with either underestimating or overestimating the error(with respect to its constant value in the plateau region). This problem shows up whenever one tries to analize statistically a too small sample of measurements.

On the other hand, 
- increasing the number of blocks outside the optimal region always results in underestimating the error. From a mathematical point of view, the decrease of the error with the number of blocks is a mere consequence of its mathematical formula. However, it is also clear that this process cannot be taken to its limit, as there is always an upper bound to the precision at which a measurement can be taken, set by the measurement protocol.

Also,
- with a fixed number of points M, the maximum possible number of blocks is - of course - M (of size 1). The data-blocking method then corresponds to a single average over M points. However, even though M might be large, this M points are not all independent (see above the correlation lenghts found in the present case) and the error is therefore underestimated. We see again that the number of blocks should not be to low. 

### B. Monte carlo metropolis and molecular dynamics (MD) calculation of the radial distribution function

```python
phases = ['solid', 'liquid', 'gas']

fig = plt.figure(figsize=(12, 4), dpi=120)
plt.suptitle("g(r) calculation via Montecarlo Metropolis", size=14)
mpl.rcParams['figure.subplot.top'] = 0.8
colors = ['blue', 'green', 'red']
plot = 1
for phase in phases:
    data = np.loadtxt("data/{}_g_r.txt".format(phase)).T
    plt.subplot(1, 3, plot);
    plt.errorbar(data[1], data[2], yerr=data[3], marker='.', color = colors[plot-1], ecolor='gray', label='metropolis')
    plt.xlabel('r')
    plt.ylabel('g(r)')
    plt.title('{} phase'.format(phase))
    plt.legend()
    plot = plot + 1
plt.show()
```

### C. Monte carlo metropolis simulation of Argon

In the last part of the exercise we simulated Argon:
($\sigma = 0.34$ nm, $\epsilon/k_B = 120$ K, $m=39.948$ amu) 

in the following conditions:
1. solid phase: $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)
2. liquid phase: $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)
3. gas phase: $\rho^\star = 0.05$, $T^\star = 1.2$ (cut-off radius: $r_c = 5.0$)

```python
#Useful constants
K=1.3806*10**(-23)   #Boltzmann constant

#Argon
sigma=0.34*10**(-9)
r=120
mass=39.948*1.66054*10**(-27)
eps=r*K

phases = ['solid', 'liquid', 'gas']

for phase in phases:
    epot = np.loadtxt("data/{}_u.txt".format(phase)).T
    pres = np.loadtxt("data/{}_p.txt".format(phase)).T
    fig = plt.figure(figsize=(12, 4), dpi=120)
    fig.suptitle("{} state".format(phase), size=14)
    plt.subplot(1, 2, 1); 
    plt.errorbar(epot[0], eps*epot[2], yerr=epot[3]*eps, color = 'b', ecolor='gray', capsize=2)
    plt.title('Potential Energy [Joule]')
    plt.xlabel('# blocks of 1000 points')
    plt.ylabel('U/N')
    plt.subplots_adjust(wspace=0.3)
    plt.subplot(1, 2, 2); 
    plt.errorbar(pres[0], pres[2]*eps/(sigma**3), yerr=pres[3]*eps/(sigma**3), color = 'b', ecolor='gray', capsize=2)
    plt.title('Pressure [Pa]')
    plt.xlabel('# blocks of 1000 points')
    plt.ylabel('P')
plt.show()
```

**NB:** 

The really interesting thing to note here is that two different algorithms (the molecular dynamics and the monte carlo metropolis), working by simulating two different statistical mechanics ensembles, give the same results. This is a consequence of the equivalence of the statistical mechanics ensembles in the thermodynamic limit.
