### Exercise 07.1

#### Useful functions

```python
# autocorrelation function
def autocorrelation(v,t):
    #denominator
    ave=0
    av2=0
    for i in range(len(v)):
        ave=ave + v[i]
        av2=av2 + v[i]*v[i]
    ave=ave/(len(v))
    av2=av2/(len(v))
    #numerator
    delta_t =len(v) -t # t_max - t
    corr1=0
    corr2=0
    corr3=0
    for i in range(delta_t):
        corr1=corr1+v[i]*v[i+t]
        corr2=corr2+v[i]
        corr3=corr3+v[i+t]
    ac = corr1/delta_t-(corr2/delta_t)*(corr3/delta_t)
    ac = ac/(av2-ave**2)

    return ac

#exponential fit
def exp_fit(x,a):
    return np.exp(-a*x)

#datablocking error
def datablocking_error(v, N):
    M = len(v)
    L = int(M/N)
    measure=np.zeros(int(N))
    measure2=np.zeros(int(N))
    vblock= np.zeros(L)
    ave = 0
    ave2=0
    err=0
    for i in range(N):
        if(i==N-1):
            vblock=v[L*i::1]
        else:
            vblock = v[L*i:L*(i+1):1]
        
        measure[i] = np.mean(vblock)
        measure2[i] = measure[i]**2
        ave = ave + measure[i]/N
        ave2 = ave2 + measure2[i]/N
    
    err=np.sqrt((ave2-ave**2)/(N-1))
    return err
```

### Solid Phase

#### Equilibration


```python
solid_noneq_instant_values = np.loadtxt("solid_noneq_instant_values.txt").T
solid_eq_instant_values = np.loadtxt("solid_eq_instant_values.txt").T
```

```python
fig = plt.figure(figsize=(12, 4), dpi=120)
plt.suptitle("Equilibration of the solid phase", size=16)
plt.subplot(1, 2, 1); 
plt.plot(solid_noneq_instant_values[0][::10], solid_noneq_instant_values[1][::10], 'b-', lw=0.5)
plt.plot(solid_eq_instant_values[0][::10]+500001, solid_eq_instant_values[1][::10], 'r-', lw=0.5, label='equilibrated configurations')
plt.semilogx()
plt.legend()
plt.ylim(-8., -7.)
plt.subplot(1, 2, 2); 
plt.plot(solid_noneq_instant_values[0][::10], solid_noneq_instant_values[2][::10], 'b-', lw=0.5)
plt.plot(solid_eq_instant_values[0][::10]+500001, solid_eq_instant_values[2][::10], 'r-', lw=0.5, label='equilibrated configurations')
plt.semilogx()
plt.ylim(5., 10.)
plt.legend()
plt.show()
```

#### Autocorrelation function computation

```python
time = np.arange(101)
ac_pot = np.zeros(101)
ac_pres= np.zeros(101)

for t in time :
    ac_pot[t] = autocorrelation(solid_eq_instant_values[1], t)
    ac_pres[t] = autocorrelation(solid_eq_instant_values[2], t)
```

```python
params_pot, cov_pot= optimize.curve_fit(exp_fit, time, ac_pot)  
params_pres, cov_pres= optimize.curve_fit(exp_fit, time, ac_pres)

fig = plt.figure(figsize=(12, 4), dpi=120)
plt.subplot(1, 2, 1);
plt.plot(time, ac_pot, color='b', marker='.', markersize=4, linestyle='none')
plt.plot(time, exp_fit(time, params_pot[0]) , 'r-', lw=2, label='exponential fit')
plt.grid(True)
plt.legend()
plt.subplot(1, 2, 2);
plt.plot(time, ac_pres, color='b', marker='.', markersize=4, linestyle='none')
plt.plot(time, exp_fit(time, params_pres[0]) , 'r-', lw=2, label='exponential fit')
plt.grid(True)
plt.legend()
plt.show()

print("Estimated correlation time for potential energy = "+str(1/params_pot[0]))
print("Estimated correlation time for pressure = "+str(1/params_pres[0]))
```

#### Data-blocking error estimation

```python
Nblocks = [100, 250, 500, 1000, 2500, 5000, 10000, 25000, 50000, 100000]
L = [5000, 2000, 1000, 500, 200, 100, 50, 20, 10, 5]
err_pot = np.zeros(len(Nblocks))
err_pres = np.zeros(len(Nblocks))
for i in range(len(Nblocks)) :
    err_pot[i] = datablocking_error(solid_eq_instant_values[1], Nblocks[i])
    err_pres[i] = datablocking_error(solid_eq_instant_values[2], Nblocks[i])

fig=plt.figure(figsize=(12, 4), dpi=120)
plt.subplot(1, 2, 1); plt.plot(L, err_pot, 'b-')
plt.grid(True)
plt.subplot(1, 2, 2); plt.plot(L, err_pres, 'b-')
plt.grid(True)
plt.show()
```
