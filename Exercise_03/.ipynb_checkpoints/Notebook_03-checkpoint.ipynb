```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("output.txt").T

plt.figure(figsize=(10, 7), dpi=100)
plt.subplot(2, 2, 1); plt.errorbar(data[0], data[1], yerr=data[2], color='b', ecolor="gray", capsize=2)
plt.grid(True)
plt.semilogx()
plt.xlabel("number of asset prices")
plt.ylabel("call option")
plt.subplot(2, 2, 2); plt.errorbar(data[0], data[3], yerr=data[4], color='b', ecolor="gray", capsize=2)
plt.grid(True)
plt.semilogx()
plt.xlabel("number of asset prices")
plt.ylabel("put option")
plt.subplot(2, 2, 3); plt.errorbar(data[0], data[5], yerr=data[6], color='b', ecolor="gray", capsize=2)
plt.grid(True)
plt.semilogx()
plt.xlabel("number of asset prices")
plt.ylabel("call option")
plt.subplot(2, 2, 4); plt.errorbar(data[0], data[7], yerr=data[8], color='b', ecolor="gray", capsize=2)
plt.grid(True)
plt.semilogx()
plt.xlabel("number of asset prices")
plt.ylabel("put option")
plt.subplots_adjust(wspace=0.3)
plt.subplots_adjust(hspace=0.8)
fig = plt.gcf()
fig.text(0.25, 0.9, "By sampling directly the final asset price S(T)", size = 16)
fig.text(0.25, 0.42, "By sampling the discretized GBM path of the asset price", size = 16)
plt.show()
```
