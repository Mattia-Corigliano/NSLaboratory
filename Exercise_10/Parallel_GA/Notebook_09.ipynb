```python
import numpy as np
import matplotlib.pyplot as plt
```

## Exercise 9.1

```python
cities_configuration = np.loadtxt("circle_cities_config.txt").T
cities_configuration2 = np.loadtxt("square_cities_config.txt").T

plt.figure(figsize=(12, 4), dpi=120)
plt.suptitle("2D configuration of the cities", size = 16)
plt.subplot(1, 2, 1);
plt.plot(cities_configuration[1], cities_configuration[2], 'bs', markersize = 6)
plt.xlabel("x", size = 12)
plt.ylabel("y", size = 12)
plt.title("Cities on a circumference")
plt.grid(True)
plt.subplot(1, 2, 2);
plt.plot(cities_configuration2[1], cities_configuration2[2], 'bs', markersize = 6)
plt.subplots_adjust(wspace=0.3)
plt.xlabel("x", size = 12)
plt.ylabel("y", size = 12)
plt.title("Cities inside a square")
plt.grid(True)
plt.show()
```

### Solution

#### Circle

```python
circle_solution = np.loadtxt("circle_opt_tour.txt").T

plt.figure(figsize=(12, 4), dpi=180)
plt.subplot(1, 2, 1);
plt.suptitle("Travelling Salesman Problem", size =16)
plt.plot(cities_configuration[1], cities_configuration[2], 'bs', markersize = 6)
plt.xlabel("x", size = 12)
plt.ylabel("y", size = 12)
plt.grid(True)
plt.subplot(1, 2, 2)
plt.plot(cities_configuration[1], cities_configuration[2], 'bs', markersize = 6)
plt.plot(circle_solution[1], circle_solution[2], 'r--', lw=2, label = "Genetic Algorithm solution")
plt.xlabel("x", size = 12)
plt.ylabel("y", size = 12)
plt.legend()
plt.grid(True)
plt.subplots_adjust(wspace = 0.3)
plt.show()
```

### Convergence

```python
tour_convergence = np.loadtxt("circle_tour_convergence.txt").T

plt.figure(figsize=(12, 6), dpi=180)
plt.suptitle("Convergence of optimal tours", size=16)
plt.subplot(2, 3, 1);
plt.plot(cities_configuration[1], cities_configuration[2], 'bs', markersize = 6)
plt.grid(True)
gen = [0, 50, 100, 150, 500]
for k in range(5):
    plt.subplot(2, 3, k+2);
    plt.plot(cities_configuration[1], cities_configuration[2], 'bs', markersize = 6)
    plt.plot(tour_convergence[1][k*33:(k+1)*33], tour_convergence[2][k*33:(k+1)*33], 'r--', label = 'gen = {}'.format(gen[k]))    
    plt.xlabel('x', size=12)
    plt.ylabel('y', size=12)
    plt.legend()
    plt.grid(True)

plt.show()
```

```python
tour_convergence = np.loadtxt("square_tour_convergence.txt").T

plt.figure(figsize=(12, 6), dpi=180)
plt.suptitle("Convergence of optimal tours", size=16)
plt.subplot(2, 3, 1);
plt.plot(cities_configuration2[1], cities_configuration2[2], 'bs', markersize = 6)
plt.grid(True)
gen = [0, 100, 200, 500, 1000]
for k in range(5):
    plt.subplot(2, 3, k+2);
    plt.plot(cities_configuration2[1], cities_configuration2[2], 'bs', markersize = 6)
    plt.plot(tour_convergence[1][k*33:(k+1)*33], tour_convergence[2][k*33:(k+1)*33], 'r--', label = 'gen = {}'.format(gen[k]))    
    plt.xlabel('x', size=12)
    plt.ylabel('y', size=12)
    plt.legend()
    plt.grid(True)

plt.show()
```
