## Metropolis evaluation of the Hydrogen atom radius

*In this computational exercise we addressed the problem of numerically estimating the expectation value of the radius of the hydrogen atom in two different electron orbital configurations: 1s and 2p. The hydrogen atom is a system pertaining to the realm of quantum mechanics, in which expectation values are calculated as integral with respect to the probability density $|\psi|^2$.*.

*We used the Metropolis algorithm to sample points from the distribution $|\psi|^2$ and then performed averages of the radius of the sampled points with the data-blocking techniques. The result of this procedure gives the wanted expected value.*

*We show our numerical results together with the analytical exact values. We also display the distribution of the sampled points, in which we can recognize the known orbitals shape.*

```python
### importing useful libraries
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

##  setting the Layout
import matplotlib as mpl
## LINES
mpl.rcParams['lines.linewidth'] = 2
mpl.rcParams['lines.linestyle'] = '-'
## AXES
mpl.rcParams['axes.linewidth'] = 1 ## edge linewidth
mpl.rcParams['axes.grid'] = True ## display grid or not
mpl.rcParams['axes.titlesize'] = 16 ## fontsize of the axes title
#mpl.rcParams['axes.titleweight'] = 'bold'
mpl.rcParams['axes.labelsize'] = 14  ## fontsize of the x any y labels
mpl.rcParams['axes.formatter.limits'] = [-2,6] # use scientific notation if log10
mpl.rcParams['axes.formatter.use_mathtext'] = True # When True, use mathtext for scientific
mpl.rcParams['legend.fontsize'] = 12
## TICKS
mpl.rcParams['xtick.labelsize'] = 12 ## fontsize of the tick labels
mpl.rcParams['ytick.labelsize'] = 12 
mpl.rcParams['xtick.minor.visible'] = False ## visibility of minor ticks on x-axis
mpl.rcParams['ytick.minor.visible'] = False
## FIGURE
mpl.rcParams['figure.titleweight'] = 'bold' ## weight of the figure title
mpl.rcParams['figure.subplot.wspace'] = 0.4
mpl.rcParams['figure.subplot.hspace'] = 0.3
## ERRORBARS
mpl.rcParams['errorbar.capsize'] = 2  ## length of end cap on error bars in pixels
```

### 1s Orbital configuration

```python
fig = plt.figure(figsize=(12, 8), dpi=100)
plt.suptitle('1s orbital configuration', size=14)
mpl.rcParams['figure.subplot.top'] = 0.9

Nplots = [1,2,3,4]
subplots = [221, 222, 223, 224]
Data = ['data/output1_uniform.txt', 'data/output1_gaussian.txt',
        'data/output3_uniform.txt', 'data/output3_gaussian.txt']
titles = ['$|\psi_{1, 0, 0}(x, y, z)|^2$ Uniform transition probability', 
          '$|\psi_{1, 0, 0}(x, y, z)|^2$ Gaussian transition probability',
          '        Uniform transition probability', 
          '        Gaussian transition probability']

for plot in Nplots:
    data = np.loadtxt(Data[plot-1]).T
    
    if(plot<3):
        x=data[1]
        y=data[2]
        z=data[3]
        ax = fig.add_subplot(subplots[plot-1], projection='3d')
        ax.scatter(x, y, z, c=z, marker='.')
        ax.set_xlabel('x')
        ax.set_ylabel('y')
        ax.set_zlabel('z')
        ax.view_init(10, 30)
    
    if(plot>=3):
        ax = fig.add_subplot(subplots[plot-1])
        block_length = int(data[0][0])
        ax.errorbar(data[0]/block_length, data[1]-1.5, yerr=data[2], color='black', ecolor="gray")
        plt.xlabel('# blocks of {} points'.format(block_length))
        plt.ylabel('<r>-3/2 in $a_0$ units')
        plt.axhline(0, ls = '--', color = 'red')
    ax.set_title(titles[plot-1])

plt.show()
```

### 2p Orbital configuration

```python
fig = plt.figure(figsize=(12, 8), dpi=100)
plt.suptitle('2p Orbital configuration', size=14)
mpl.rcParams['figure.subplot.top'] = 0.9

Nplots = [1,2,3,4]
subplots = [221, 222, 223, 224]
Data = ['data/output2_uniform.txt', 'data/output2_gaussian.txt',
        'data/output4_uniform.txt', 'data/output4_gaussian.txt']
titles = ['$|\psi_{2, 1, 0}(x, y, z)|^2$ Uniform transition probability', 
          '$|\psi_{2, 1, 0}(x, y, z)|^2$ Gaussian transition probability',
          '   Uniform transition probability', 
          '   Gaussian transition probability']

for plot in Nplots:
    data = np.loadtxt(Data[plot-1]).T
    
    if(plot<3):
        x=data[1]
        y=data[2]
        z=data[3]
        ax = fig.add_subplot(subplots[plot-1], projection='3d')
        ax.scatter(x, y, z, c=z, marker='.')
        ax.set_xlabel('x')
        ax.set_ylabel('y')
        ax.set_zlabel('z')
        ax.view_init(10, 30)
    
    if(plot>=3):
        block_length = int(data[0][0])
        ax = fig.add_subplot(subplots[plot-1])
        ax.errorbar(data[0]/block_length, data[1]-5, yerr=data[2], color='black', ecolor="gray")
        plt.xlabel('# blocks of {} points'.format(block_length))
        plt.ylabel('<r>-5 in $a_0$ units')
        plt.axhline(0, ls = '--', color = 'red')
        
    ax.set_title(titles[plot-1])

plt.show()
```

**NB:**

- How large should be the number of Monte Carlo Metropolis steps in each block?

As said in the Notebook01, much care should be used in the choice of the number of blocks, and therefore of the number of points inside each block, for the data-blocking technique. Indeed, even though this choice does not alter the final result (averaging N averages over M points is the same as to average once over NM points), a bad tradeoff between the number of blocks and their sizes could affect the reliability of the estimate of the errors.

Indeed (here we consider fixed total number of points), for example, choosing too much blocks could result in an underestimate of the error in the result.

**Note** that this might be what had happened with the Gaussian transition probabilities in the previous panels.
--> The apparent incompatibility of the results with the expected values might be the reflection of the usage of too much blocks (1000).

On the other hand, there is also a lower bound in the number of blocks to be used. Indeed, one must always make sure that the number of blocks is large enough to ensure that a statistical treatment of the error makes sense (it has no sense to use 2 blocks and compute the standard deviation).

Therefore, we can conclude that lowering a bit (maybe 100 instead of 1000) the number of blocks to be used in the present analysis is a good idea for a more reliable estimate of the errors in the results.

- The code performs an internal optimization routine of the transition probability step, which is choose in order to obtain 50% of acceptance.


- In the results shown in the panels above, the origin was chosen as the starting point of the simulation. 

In the end, we also run a simulation starting far from the origin.


**Starting from a point far from the origin**

```python
fig = plt.figure(figsize=(12, 8), dpi=100)
plt.suptitle('starting far from origin', size=14)
mpl.rcParams['figure.subplot.top'] = 0.9

Nplots = [1,2,3,4]
subplots = [221, 222, 223, 224]
Data = ['data/far_from_origin/output_uniform.txt', 'data/far_from_origin/output2_uniform.txt',
        'data/far_from_origin/output3_uniform.txt', 'data/far_from_origin/output4_uniform.txt']
titles = ['$|\psi_{1, 0, 0}(x, y, z)|^2$ Uniform transition probability', 
          '$|\psi_{2, 1, 0}(x, y, z)|^2$ Uniform transition probability',
          '        1s', 
          '        2p']
exact_value = 1.5
for plot in Nplots:
    data = np.loadtxt(Data[plot-1]).T
    
    if(plot<3):
        x=data[1]
        y=data[2]
        z=data[3]
        ax = fig.add_subplot(subplots[plot-1], projection='3d')
        ax.scatter(x, y, z, c=z, marker='.')
        ax.set_xlabel('x')
        ax.set_ylabel('y')
        ax.set_zlabel('z')
        ax.view_init(10, 30)
    
    if(plot>=3):
        ax = fig.add_subplot(subplots[plot-1])
        block_length = int(data[0][0])
        ax.errorbar(data[0]/block_length, data[1]-exact_value, yerr=data[2], color='black', ecolor="gray")
        plt.xlabel('# blocks of {} points'.format(block_length))
        plt.ylabel('<r>-{} in $a_0$ units'.format(exact_value))
        exact_value = 5
        plt.axhline(0, ls = '--', color = 'red')
    ax.set_title(titles[plot-1])

plt.show()
```
