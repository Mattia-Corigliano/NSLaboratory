# Sampling the Hydrogen orbitals wave functions with Metropolis algorithm

### 1s Orbital configuration

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

data = np.loadtxt("output_uniform.txt").T
data2 = np.loadtxt("output_gaussian.txt").T
data3 = np.loadtxt("output3_uniform.txt").T
data4 = np.loadtxt("output3_gaussian.txt").T
x=data[1]
y=data[2]
z=data[3]
x2=data2[1]
y2=data2[2]
z2=data2[3]
fig = plt.figure(figsize=(12, 8), dpi=100)
ax = fig.add_subplot(221, projection='3d')
#ax = Axes3D(fig)
ax.scatter(x, y, z, c=z, marker='.')
ax.set_title("Uniform transition probability")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
#ax2 = Axes3D(fig2)
ax = fig.add_subplot(222, projection='3d')
ax.scatter(x2, y2, z2, c=z2, marker='.')
ax.set_title("Gaussian Transition probability")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)

ax = fig.add_subplot(223)
ax.errorbar(data3[0], data3[1], yerr=data3[2], color='b', ecolor="gray", capsize=2)
plt.grid(True)
ax = fig.add_subplot(224)
ax.errorbar(data4[0], data4[1], yerr=data4[2], color='b', ecolor="gray", capsize=2)
plt.grid(True)
plt.show()
```
